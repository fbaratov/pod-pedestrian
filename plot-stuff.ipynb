```python
import json
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.backends.backend_pdf import PdfPages
```

```python
data = json.load(open('train.json','r'))
```

```python
data.keys()
```

```python
loss = data['loss']
val_loss = data['val_loss']
```

```python
plt.plot(loss, label="Training loss")
plt.plot(val_loss, label="Validation loss")
plt.legend()
plt.title("Model loss during training")
plt.ylabel("Model loss")
plt.xlabel("Epoch")
plt.savefig("figures/model-training.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

```python
plt.plot(data['localization'], label="Training localization")
plt.plot(data['val_localization'], label="Validation localization")
plt.legend()
plt.title("Model localization loss during training")
plt.ylabel("Model localization loss")
plt.xlabel("Epoch")
plt.show()
```

```python
plt.plot(data['negative_classification'], label="Training negative classification")
plt.plot(data['val_negative_classification'], label="Validation negative classification")
plt.legend()
plt.title("Model negative classification loss during training")
plt.ylabel("Model negative classification loss")
plt.xlabel("Epoch")
plt.show()
```

```python
plt.plot(data['negative_classification'], label="Negative classification loss", linestyle='dashed')
plt.plot(data['positive_classification'], label="Positive classification loss", linestyle='dashed')
classification = [data['negative_classification'][i]+data['positive_classification'][i] for i in range(len(data['negative_classification']))]
plt.plot(classification, label="Total classification loss")
plt.plot(data['localization'], label="Localization loss")


plt.legend()
plt.title("Breakdown of training loss")
plt.ylabel("Loss value")
plt.xlabel("Epoch")
plt.savefig("figures/multibox-breakdown.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

```python
plt.plot(data['val_negative_classification'], label="Validation negative classification loss")
plt.plot(data['val_positive_classification'], label="Validation positive classification loss")
plt.plot(data['val_localization'], label="Training localization loss")

plt.legend()
plt.title("Model loss during validation")
plt.ylabel("Loss value")
plt.xlabel("Epoch")
plt.show()
```

```python
plt.plot(data['positive_classification'], label="Training positive classification")
plt.plot(data['val_positive_classification'], label="Validation positive classification")
plt.legend()
plt.title("Model positive classification loss during training")
plt.ylabel("Model positive classification loss")
plt.xlabel("Epoch")
plt.show()
```
